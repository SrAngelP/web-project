```python
url_doc = requests.get(link_xml);

soup = BeautifulSoup(url_doc.content, 'xml');
date = soup.select('pub-date',{'pub-type':'epub'})
date = (date[1].day.extract()).get_text() + '/' + (date[1].month.extract()).get_text()  + '/' + (date[1].year.extract()).get_text()
text = soup.select('body')[0]
#pattern = r'\[[0-9+0-9*]+\]+'
#text = re.sub(pattern, '', text)
#pattern2 = r'\(Fig. 0-9*A-Za-z*\)'
#re.sub(pattern, '', text)
while text.find('xref') != None:
    text.xref.extract()


figs = []
while text.find('fig') != None:
    figs.append(text.fig.extract())
    
text = text.get_text()
words = [re.split(' ', x) for x in re.split('\n', text)]
words = [(x.strip('()""-.,')).lower() for lista in words for x in lista if (len(x)>1)&((x.strip('()""-.,')).isalpha()==True)]
#words = list(set(words))
print(len(words))
df_words = pd.DataFrame(words, columns = ['Words'])
((df_words.groupby(['Words']).agg({'Words':'count'})).add_suffix('_Count').reset_index()).sort_values('Words_Count', ascending=False)
```
