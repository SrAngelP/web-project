```python
response = requests.get(url)
results = response.json()
data = pd.json_normalize(results)
data = pd.DataFrame(data['value'][0])

countries = pd.read_json('countries.json')


for i in range(len(countries)):
    data.loc[data['SpatialDim']==countries[0][i], ['SpatialDim']] = countries[1][i]       #Est√° leyendo un DataFrame por eso el 1 y 0 van primero

env_causes = [['ENVCAUSE000','Total'], 
              ['ENVCAUSE039', 'Lower respiratory infections'], 
              ['ENVCAUSE068', 'Trachea, bronchus, lung cancers'], 
              ['ENVCAUSE113', 'Ischaemic heart disease'], 
              ['ENVCAUSE114','Stroke'], 
              ['ENVCAUSE118', 'Chronic obstructive pulmonary disease']]

for i in range(len(env_causes)):
    data.loc[data['Dim2']==env_causes[i][0], ['Dim3Type']] = env_causes[i][0]          #Leyendo una lista



for i in range(len(env_causes)):
    data.loc[data['Dim2']==env_causes[i][0], ['Dim3']] = env_causes[i][1]



for i in range(len(env_causes)):
    data.loc[data['Dim3']==env_causes[i][0], ['Dim3']] = env_causes[i][1]



rg_nms = [['EMR','Eastern Mediterranean'], 
              ['AFR', 'Africa'], 
              ['AMR', 'Americas'], 
              ['EUR', 'Europe'], 
              ['SEAR','South-East Asia'], 
              ['WPR', 'Western Pacific']]

for i in range(len(rg_nms)):
    data.loc[data['SpatialDim']==rg_nms[i][0], ['SpatialDim']] = rg_nms[i][1]



unrg_nms = [['9','Oceania'],
            ['15', 'Northern Africa'],
            ['21', 'Northern America'], 
            ['30', 'Eastern Asia'],
            ['34','Southern Asia'],
            ['35', 'South-eastern Asia'], 
            ['143', 'Central Asia'], 
            ['145', 'Western Asia'], 
            ['150', 'Europe'], 
            ['202','Sub-Saharan Africa'], 
            ['419', 'Latin America and the Caribbean']];

for i in range(len(unrg_nms)):
    data.loc[data['SpatialDim']== unrg_nms[i][0], ['SpatialDim']] = unrg_nms[i][1]




unsdgrg_nms = [['UNSDG_AUSTRALIANEWZEALAND','Australia and New Zealand'],
            ['UNSDG_CENTRALASIASOUTHERNASIA', 'Central Asia and Southern Asia'],
            ['UNSDG_EASTERNASIASOUTHEASTERNASIA', 'Eastern Asia and South-eastern Asia'], 
            ['UNSDG_EUROPENORTHERNAMR', 'Europe and Northern America'],
            ['UNSDG_LAC','Latin America and the Caribbean'],
            ['UNSDG_OCEANIAexAUSNZL', 'Oceania excluding Australia and New Zealand'], 
            ['UNSDG_SUBSAHARANAFRICA', 'Sub-Saharan Africa'], 
            ['UNSDG_WESTERNASIANORTHERNAFR', 'Western Asia and Northern Africa']];

for i in range(len(unsdgrg_nms)):
    data.loc[data['SpatialDim']== unsdgrg_nms[i][0], ['SpatialDim']] = unsdgrg_nms[i][1]
    
range_ages = [['AGEAll','All'], ['AGE25-29', '25-29'], ['AGE30-34', '30-34'],
       ['AGE35-39', '35-39'], ['AGE40-44', '40-44'], ['AGE45-49', '45-49'], ['AGE50-54', '50-54'], ['AGE55-59', '55-59'],
       ['AGE60-64', '60-64'], ['AGE65-69', '65-69'], ['AGE70-74', '70-74'], ['AGE75-79', '75-79'], ['AGE80+', '80+'], 
        ['AGE0-4', '0-4'], ['AGE5-9', '5-9'], ['AGE10-14', '10-14'], ['AGE15-19', '15-19'], ['AGE20-24', '20-24']]

for i in range(len(range_ages)):
    data.loc[data['Dim2']== range_ages[i][0], ['Dim2']] = range_ages[i][1]


data.loc[data['Dim1']=='BTSX', ['Dim1']] = 'Both'
data.loc[data['Dim1']=='FMLE', ['Dim1']] = 'Female'
data.loc[data['Dim1']=='MLE', ['Dim1']] = 'Male'

data.loc[data['SpatialDim']=='WB_LIMI', ['SpatialDim']] = 'Low- and middle-income'
data.loc[data['SpatialDim']=='WB_HI', ['SpatialDim']] = 'High-income'



len(data)
```

```python
deaths_ctry = data[data['SpatialDimType']=='COUNTRY']
#deaths_ctry = data[data['Dim2Type']=='ENVCAUSE']
deaths_ctry = ((deaths_ctry.loc[:,'SpatialDim':'Date']).dropna(axis=1,how='all').drop(columns=['TimeDimType', 'TimeDim', 'Date', 'Value', 'Dim2Type', 'Dim2', 'Dim1Type', 'Dim3Type'])).drop(deaths_ctry[deaths_ctry['Dim3']=='Total'].index)
deaths_ctry = deaths_ctry.rename(columns={'SpatialDim':'Country', 'Dim1':'Sex', 'Dim3':'Cause', 'NumericValue':'Value', 'Low':'Low', 'High':'High'})
country_names = pd.unique(deaths_ctry['Country'])
deaths_country = {}
for i in country_names:
    deaths_country[i] = ((deaths_ctry[deaths_ctry['Country']==i].loc[:,'Sex':'High']).reset_index(drop=True))
    


```
